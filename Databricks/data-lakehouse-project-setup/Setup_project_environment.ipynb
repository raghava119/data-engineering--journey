# set up project environment for the circuitBox data lakehouse 
- 1.set up the external location dea_course_ext_dl_circuitbox
- 2.creat catalog -- circuitbox
- 3.create schemas --- landing, lakehouse(bronze, silver,gold)
- 4.create volume --operational data 

1.Create External location


```python
%sql
Create external location if not exists dea_course_ext_dl_circuitbox
URL 'abfss://circiuitbox@deacourseextdly.dfs.core.windows.net/'
WITH (storage credential `dea-course-ext-sc`)
comment 'external location for circuitbox'
```

```python
%fs ls  'abfss://circiuitbox@deacourseextdly.dfs.core.windows.net/'
```

2.Create Catalog

```python
%sql
create catalog if not exists circuitbox
managed location 'abfss://circiuitbox@deacourseextdly.dfs.core.windows.net/'
comment 'catalog for the circuitbox data lakehouse'
```

```python
%sql show catalogs;
```

```python
%sql
create schema if not exists circuitbox.landing 
managed location 'abfss://circiuitbox@deacourseextdly.dfs.core.windows.net/landing'
comment 'landing zone for circuitbox data';

create schema if not exists circuitbox.lakehouse 
managed location 'abfss://circiuitbox@deacourseextdly.dfs.core.windows.net/lakehouse'
comment 'landing zone for circuitbox data'
```

```python
%sql
use  catalog circuitbox;
show schemas;
```

4.create volumes

```python
%sql
use  catalog circuitbox;
use schema landing;
create external volume if not exists operational_data
location 'abfss://circiuitbox@deacourseextdly.dfs.core.windows.net/landing/operationaldata';
```
